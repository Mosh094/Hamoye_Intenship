Loading Fuel Data

```python
url = 'https://github.com/WalePhenomenon/climate_change/blob/master/fuel_ferc1.csv?raw=true'
fuel_data = pd.read_csv(url, error_bad_lines=False)
fuel_data.describe(include='all')
```

```python
fuel_data
```

```python
#Check missing items
fuel_data.isnull()
```

```python
#Check number of missing items per column
fuel_data.isnull().sum()
```

```python
#A dataframe can be easily categorised into different segments based on a given criteria using the groupby() function. 
#This initially splits the dataframe into the groups then applies a function to the groups after which the results are combined.
#use groupby to count the sum of each unique value in the fuel unit column
fuel_data.groupby('fuel_unit')['fuel_unit'].count()

```

```python
#Fill fields with missing items with mcf
fuel_data[['fuel_unit']] = fuel_data[['fuel_unit']].fillna(value='mcf')
```

```python
#check if missing values have been filled
fuel_data.isnull().sum()
```

```python
fuel_data.groupby('report_year')['report_year'].count()
```

```python
#group by the fuel type code year and print the first entries in all the groups formed
fuel_data.groupby('fuel_type_code_pudl').first()
```

To show how to merger data, first split into fuel_df1 and fuel_df2

```python
fuel_df1 = fuel_data.iloc[0:19000].reset_index(drop=True) # pick from row 0 to row 18999 i.e beggining to 18999
fuel_df2 = fuel_data.iloc[19000:].reset_index(drop=True) # pick from row 18999 to the final row
```

```python
fuel_df1
```

```python
fuel_df2
```

```python
#check that the length of both dataframes sum to the expected length
assert len(fuel_data) == (len(fuel_df1) + len(fuel_df2))
```

```python
#an inner merge will lose rows that do not match in both dataframes
pd.merge(fuel_df1, fuel_df2, how="inner")
```

```python
#outer merge returns all rows in both dataframes
pd.merge(fuel_df1, fuel_df2, how="outer")
```

```python
#removes rows from the right dataframe that do not have a match with the left
#and keeps all rows from the left
pd.merge(fuel_df1, fuel_df2, how="left")
```

```python
#removes rows from the left dataframe that do not have a match with the right
#and keeps all rows from the right
pd.merge(fuel_df1, fuel_df2, how="right")
```

```python
#check for duplicate rows
fuel_data.duplicated().any()
```

To concat, import sample data saved as data_to_concat

```python
csv_df = pd.read_csv('data_to_concat.csv')
```

```python
pd.concat([fuel_data, csv_df]).reset_index(drop=True)
```

# Lesson 5 -  Data Visualisation

```python
# Import plotting library
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(7,4))
plt.xticks(rotation=90)
fuel_unit = pd.DataFrame({'unit':['BBL', 'GAL', 'GRAMSU', 'KGU', 'MCF', 'MMBTU', 'MWDTH', 'MWHTH', 'TON'],
            'count':[7998, 84, 464, 110, 11354, 180, 95, 100, 8958]})
sns.barplot(data=fuel_unit, x='unit', y='count')
plt.xlabel('Fuel Unit')
```

```python
#Because of the extreme range of the values for the fuel unit, we can plot the barchart by taking the logarithm of the y-axis as follows:


g = sns.barplot(data=fuel_unit, x='unit', y='count')
g.set_yscale("log")
g.set_ylim(1, 12000)
plt.xlabel('Fuel Unit')
```

```python
# Select a sample of the dataset
sample_df = fuel_data.sample(n=50, random_state=4)
sns.regplot(x=sample_df["utility_id_ferc1"], y=sample_df["fuel_cost_per_mmbtu"], fit_reg=False)
```

```python
# Import plotting library
import seaborn as sns

# Box plot
sns.boxplot(x="fuel_type_code_pudl", y="utility_id_ferc1",
            palette=["m", "g"], data=fuel_data)
# KDE plot 
sns.kdeplot(sample_df['fuel_cost_per_unit_burned'], shade=True, color="b")
```
